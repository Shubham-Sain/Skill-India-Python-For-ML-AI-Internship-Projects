```python
#IMPORTING ALL THE LIBRARIES AND FUNCTIONS NEEDED FOR THE WORKING OF THE NEURAL NETWORK MODEL
import pandas as pd
import numpy as np

from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense
```

```python
#STORING THE WHOLE DATABASE OF TUMOR TEST CASES IN 'data'
data = load_breast_cancer()
```

```python
#POSSIBLE TYPES OF TUMOR
data['target_names']
```

```python
#ASSIGNING FEATURES AND LABELS FOR THE TUMOR DATABASE
feature = data['data']
label = data['target']
```

```python
#SIZE OF FEATURES LIST OF THE DATABASE
feature.shape
```

```python
#SIZE OF LABELS LIST OF THE DATABASE
label.shape
```

```python
#STANDARDIZING ALL THE DETAILS OF THE DATABASE
scale = StandardScaler()

feature = scale.fit_transform(feature)
```

```python
#FEATURES OF THE WHOLE DATABASE WITH STANDARDIZED DETAILS SHOWN IN A DATA FRAME
df_feature = pd.DataFrame(feature, columns = data['feature_names'])
df_feature
```

```python
#PRINTING ALL THE STANDARDIZED FEATURES ABOUT ANY TUMOR CASE IN THE DATABASE
i = int(input("Enter the serial number of the tumor in the database whose all standardized details are to be shown : "))
print()
j = 0

for k in data['feature_names']:
  print(k, " :  ", feature[i][j])
  j += 1
```

```python
#LABELS OF THE WHOLE DATABASE WITH STANDARDIZED FEATURES SHOWN IN A DATA FRAME
df_label = pd.DataFrame(label, columns = ['label'])
df_label
```

```python
#CONCATENATING THE DATAFRAMES OF FEATURES AND LABELS OF THE WHOLE DATABASE WITH STANDARDIZED FEATURES IN A SINGLE DATA FRAME
df = pd.concat([df_feature, df_label], axis=1)
df
```

```python
#SHUFFLING THE ROWS OF THE CONCATENATED DATA FRAME
df = df.sample(frac = 1)
df
```

```python
#DIVIDING THE CONCATENATED DATA FRAME INTO 'feature' AND 'label' LISTS
feature = df.values[ : , :30 ]
label = df.values[ : , 30: ]
```

```python
#DIVIDING 569 TUMORS INTO TRAINING, VALIDATION & TESTING â€”

#500 TUMORS â€” Training
X_training = feature[:500]
y_training = label[:500]


#35 TUMORS  â€” Validation
X_validation = feature[500:535]
y_validation = label[500:535]


#34 TUMORS  â€” Testing
X_testing = feature[535:]
y_testing = label[535:]
```

```python
#ESTABLISHING A NEURAL NETWORK MODEL FOR THE PREDICTION OF THE TYPE OF THE TUMORS
model = Sequential()

model.add(Dense(30, activation = 'relu', input_dim = 30))
model.add(Dense(15, activation = 'relu'))
model.add(Dense(8, activation = 'relu'))
model.add(Dense(4, activation = 'relu'))
model.add(Dense(2, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))

model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
#TRAINING THE NEURAL NETWORK MODEL WITH A DATABASE OF FIRST 500 TUMOR TEST CASES
#AND VALIDATING THE NEURAL NETWORK MODEL WITH A DATABASE OF NEXT 35 UNTESTED-BEFORE TUMOR TEST CASES
model.fit(X_training, y_training,batch_size = 3, epochs = 5, validation_data = (X_validation, y_validation))
```

```python
#FINAL TESTING THE NEURAL NETWORK MODEL WITH A DATABASE OF THE LAST 34 UNTESTED-BEFORE TUMOR TEST CASES
model.evaluate(X_testing, y_testing)
```

```python
#PRINTING THE RESULTS OF THE PREDICTION OF TYPE OF THE TUMOR FOR THE LAST 34 TUMOR TEST CASES
#AND PRINTING THEIR ACTUAL TYPE ALREADY PROVIDED IN THE DATABASE TO MATCH THE RESULTS OF THE NEURAL NETWORK MODEL
for i in range(34):
  sample = X_testing[i]
  sample = np.reshape(sample, (1,30))
  
  print("SERIAL NO. OF TEST CASE IN 'testing' list: ",i+1)

  if(model.predict(sample)[0][0] > 0.5):
    print("PREDICTED TYPE OF TUMOR BY NEURAL NETWORK MODEL â€” Benign")
  else:
    print("PREDICTED TYPE OF TUMOR BY NEURAL NETWORK MODEL â€” Malignant")

  if(y_testing[i] == 1):
    print("ACTUAL TYPE OF TUMOR IN THE DOWNLOADED DATABASE â€” Benign")
  else:
    print("ACTUAL TYPE OF TUMOR IN THE DOWNLOADED DATABASE â€” Malignant")
  
  print("\n---------------------------------------\n")
```
